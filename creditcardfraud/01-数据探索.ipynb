```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

data = pd.read_csv('./dataset/creditcard.csv')
```

## 数据探索

```python
#  观察样本的分布规则
count_classes = pd.value_counts(data['Class'],sort=True)
print(count_classes)  # 0表示为欺诈 数据284315条   1 表示欺诈  数据有492条
count_classes.plot(kind='bar')
plt.title("Fraud class histogram")
plt.xlabel("Class")
plt.ylabel("Frequency")
plt.savefig('./result_img/data_ditribution.png',bbox_inches="tight")
plt.show()
```

```python
# 欺诈交易随时间的分布情况
bins = 50
plt.hist(data.Time[data.Class==1], bins=bins)
plt.title("Fraud")
plt.xlabel('Time')
plt.ylabel('Number of Transactions')
plt.savefig('./result_img/fraud_time.png',bbox_inches="tight")
plt.show()



plt.hist(data.Time[data.Class==0], bins=bins)
plt.title("Normal")
plt.xlabel('Time')
plt.ylabel('Number of Transactions')
plt.savefig('./result_img/normal_time.png',bbox_inches="tight")
plt.show()
```

```python
# 交易金额分布情况
bins = 30
plt.hist(data.Amount[data.Class==1], bins=bins)
plt.title("Fraud")
plt.xlabel("Amount($)")
plt.ylabel("Number of Transcations")
plt.yscale("log")
plt.savefig('./result_img/fraud_amount.png',bbox_inches="tight")
plt.show()

plt.hist(data.Amount[data.Class==0], bins=bins)
plt.title("Normal")
plt.xlabel("Amount($)")
plt.ylabel("Number of Transcations")
plt.yscale("log")
plt.savefig('./result_img/normal_amount.png',bbox_inches="tight")
plt.show()
```
